## Import libraries

```python
import logging
import os
import tkinter as tk
from collections import Counter
from tkinter import filedialog

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
from PIL import Image
from sklearn.metrics import (
    ConfusionMatrixDisplay,
    accuracy_score,
    auc,
    classification_report,
    confusion_matrix,
    f1_score,
    precision_score,
    recall_score,
    roc_auc_score,
    roc_curve,
)
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from tensorflow.keras import layers, models
from tensorflow.keras.applications import EfficientNetB3
from tensorflow.keras.callbacks import EarlyStopping
```

## Set Constants

```python
IMG_SIZE = 256
BATCH_SIZE = 32
EPOCHS = 10
```

## Import data
#### https://www.kaggle.com/datasets/vencerlanz09/plastic-paper-garbage-bag-synthetic-images

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "INPUT/Bag Classes/",
    seed=42,
    shuffle=True,
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
)
```

```python
classes = dataset.class_names
print(classes)
```

## Class Distribution

```python
label_counts = Counter()

for images, labels in dataset.unbatch():
    label_counts[int(labels.numpy())] += 1

class_labels = [classes[i] for i in label_counts.keys()]
class_values = list(label_counts.values())

plt.figure(figsize=(8, 5))
plt.bar(class_labels, class_values)
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Number of Images")
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Display samples of images!

```python
plt.figure(figsize=(10, 10))
plt.suptitle('Samples of Bags', fontsize=16)
plt.subplots_adjust(top=0.95)

for images, labels in dataset.take(1):
    for i in range(12):
        ax = plt.subplot(3, 4, i + 1)
        ax.imshow(images[i].numpy().astype('uint8'))
        ax.set_title(classes[labels[i]], fontsize=10)
        ax.axis('off')
```

## Split train, val, test

```python
dataset = dataset.shuffle(1000, seed=12)
total_batches = len(dataset)
train_size = int(0.8 * total_batches)
val_size = int(0.1 * total_batches)

train = dataset.take(train_size)
val = dataset.skip(train_size).take(val_size)
test = dataset.skip(train_size + val_size)
```

```python
def count_images(dataset):
    return sum([len(batch[0]) for batch in dataset])

print("Train images:", count_images(train))
print("Validation images:", count_images(val))
print("Test images:", count_images(test))
```

## Cache, Shuffle, and Prefetch the Dataset

```python
train_ds = train.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds = val.cache().prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds = test.cache().prefetch(buffer_size=tf.data.AUTOTUNE)
```

## Creating a Layer for Resizing and Normalization

```python
resize_and_scaling = tf.keras.Sequential([
    layers.Resizing(IMG_SIZE, IMG_SIZE), 
    layers.Rescaling(1.0 / 255)]
)
```

## Functions

```python
def plot_error_accuracy(history):
    plt.figure(figsize=(12, 6))

    # Plot error
    plt.subplot(1, 2, 1)
    plt.plot(history.history["loss"], label="loss")
    plt.plot(history.history["val_loss"], label="val_loss")
    plt.xlabel("Epoch")
    plt.ylabel("Error")
    plt.legend(loc="lower right")
    plt.title("Epoch vs Error")

    # Plot accuracy
    plt.subplot(1, 2, 2)
    plt.plot(history.history["accuracy"], label="accuracy")
    plt.plot(history.history["val_accuracy"], label="val_accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend(loc="lower right")
    plt.title("Epoch vs Accuracy")

    plt.tight_layout()
    plt.show()


def get_y_pred(model, test_ds):
    y_pred = model.predict(test_ds)
    y_pred_class = np.argmax(y_pred, axis=1)
    return y_pred, y_pred_class


def model_assessment(y_true, y_pred_class, classes):
    accuracy = accuracy_score(y_true, y_pred_class)
    precision_measure = precision_score(y_true, y_pred_class, average="weighted")
    recall_measure = recall_score(y_true, y_pred_class, average="weighted")
    f1 = f1_score(y_true, y_pred_class, average="weighted")

    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision_measure:.4f}")
    print(f"Recall: {recall_measure:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print("\n\n")
    report = classification_report(
        y_true, y_pred_class, target_names=classes, output_dict=True
    )
    df = pd.DataFrame(report).transpose()
    print(df)
    return df, {accuracy, precision_measure, recall_measure, f1}


def display_confusion_matrix(y_true, y_pred_class, classes):
    cm = confusion_matrix(y_true, y_pred_class)

    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)
    disp.plot(cmap=plt.cm.Blues, values_format="g")
    disp.ax_.set_title("Confusion Matrix")
    disp.ax_.set_xlabel("Predicted Label")
    disp.ax_.set_ylabel("True Label")
    plt.show()


def plot_multiclass_roc(y_true, y_pred_prob, class_names):
    y_true_bin = label_binarize(y_true, classes=np.arange(len(class_names)))

    fpr = {}
    tpr = {}
    roc_auc = {}

    for i in range(len(class_names)):
        fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_pred_prob[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    plt.figure(figsize=(10, 8))

    for i in range(len(class_names)):
        plt.plot(
            fpr[i],
            tpr[i],
            label=f"ROC for class {class_names[i]} (AUC = {roc_auc[i]:.5f})",
        )

    plt.plot([0, 1], [0, 1], "k--")
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("Multiclass ROC Curve")
    plt.legend(loc="lower right")
    plt.grid()
    plt.show()

def show_predictions(y_pred_class, samples, class_names):
    images = np.array([img for img, _ in samples])
    labels = np.array([label for _, label in samples])

    plt.figure(figsize=(20, 10))
    plt.suptitle('Sample prediction', fontsize=16)
    for i in range(20):
        plt.subplot(4, 5, i + 1)
        plt.imshow(images[i].astype("uint8"))
        plt.axis('off')
        true_label = class_names[labels[i]]
        pred_label = class_names[y_pred_class[i]]
        color = "green" if true_label == pred_label else "red"
        plt.title(f"True: {true_label}\nPred: {pred_label}", color=color, fontsize=9)
    plt.tight_layout()
    plt.show()
```

## Test Dataset

```python
y_true = tf.concat([label for _, label in test_ds], axis=0).numpy()

# Sample to show predictions
unbatched = test_ds.unbatch()
samples = list(unbatched.take(20).as_numpy_iterator())
```

## CNN Model

```python
cnn_model = models.Sequential([
    resize_and_scaling,
    
    layers.Conv2D(16, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),

    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(len(classes), activation='softmax')
])

cnn_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

early_stop = EarlyStopping(
    monitor='loss',
    patience=3,
    restore_best_weights=True
)

history = cnn_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=[early_stop]
)
```

```python
cnn_model.summary()
```

```python
plot_error_accuracy(history)
```

```python
cnn_y_pred, cnn_y_pred_class = get_y_pred(cnn_model, test_ds)

cnn_model_assessment, metrics_dict = model_assessment(y_true, cnn_y_pred_class, classes)
```

```python
display_confusion_matrix(y_true, cnn_y_pred_class, classes)
```

```python
plot_multiclass_roc(y_true, cnn_y_pred, classes)
```

```python
show_predictions(cnn_y_pred_class, samples, classes)
```

## Real-world Testing

```python
def browse_image():
    root = tk.Tk()
    root.withdraw()
    file_path = filedialog.askopenfilename(filetypes=[("Image files", "*.jpg *.png *.jpeg *.bmp")])
    return file_path
```

```python
def process_and_predict_from_path(image_path, model, class_names):
    image = Image.open(image_path).convert('RGB')
    plt.imshow(image)
    plt.axis('off')
    plt.title("Selected Image")
    plt.show()

    img_array = tf.keras.utils.img_to_array(image)
    img_array = tf.expand_dims(img_array, 0)

    predictions = model.predict(img_array)
    predicted_class = class_names[np.argmax(predictions)]
    confidence = np.max(predictions)

    print(f"Prediction: {predicted_class} (Confidence: {confidence:.4f})")
```

```python
image_path = browse_image()
process_and_predict_from_path(image_path, model=cnn_model, class_names=classes)
```
